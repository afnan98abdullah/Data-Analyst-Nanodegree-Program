# 18 cleaning practice

```python
import pandas as pd
df=pd.read_csv(r'C:\Users\HP\.jupyter\data.csv')
df.info()
```

```python
# use means to fill in missing values
mean1=df["texture_mean"].mean()
df['texture_mean'].fillna(mean1,inplace=True)

mean2=df["smoothness_mean"].mean()
df['smoothness_mean'].fillna(mean2,inplace=True)

mean3=df["symmetry_mean"].mean()
df['symmetry_mean'].fillna(mean3,inplace=True)
# confirm your correction with info()
df.info()
```

```python
# check for duplicates in the data
df.duplicated()
```

```python
# drop duplicates
df.drop_duplicates(inplace=True)
```

```python
new_labels=[]
for col in df.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])
    else:
        new_labels.append(col)
        
new_labels        
```

```python
df.columns=new_labels
```

```python
df.to_csv(r'C:\Users\HP\.jupyter\data.csv',index=False)
```
